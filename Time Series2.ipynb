```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
import matplotlib
matplotlib.rc('xtick',labelsize=1)
matplotlib.rc('ytick',labelsize=30)
```

```python
import seaborn as sns
```

```python
sns.set(style="whitegrid", color_codes=True)
from statsmodels.tsa.stattools import acf,pacf
```

```python
axisline= np.linspace(0,10,500)
#normal distributed values
normalnumber=np.random.normal(0,5,500)
```

```python
#exponential series to get the trend
ExponentialData =np.exp(axisline**0.5)
#adding normally distrributed series to exponetial series
y=normalnumber+ExponentialData
```

```python
plt.figure(figsize=(16,7))
plt.plot(y)
```

```python
ACF_lag= acf(y,nlags=200)
```

```python
plt.figure(figsize=(16,7))
plt.plot(ACF_lag,marker='+')
plt.axhline(y=0,linestyle='--',color='grey')
plt.axhline(y=-1.96/np.sqrt(len(y)),linestyle='--',color='grey')
plt.axhline(y=1.96/np.sqrt(len(y)),linestyle='--',color='grey')
plt.title('Autocorrection')
plt.xlabel('#lags')
plt.ylabel('correlation')
plt.tight_layout()
```

```python
#calling partial correlation function
PACF_lag=pacf(y,nlags=30,method='ols')
plt.figure(figsize=(16,7))
plt.plot(PACF_lag,marker='+')
plt.axhline(y=0,linestyle='--',color='grey')
plt.axhline(y=-1.96/np.sqrt(len(y)),linestyle='--',color='grey')
plt.axhline(y=1.96/np.sqrt(len(y)),linestyle='--',color='grey')
plt.title('Partial Autocorrection')
plt.xlabel('#lags')
plt.ylabel('correlation')
plt.tight_layout()
```
